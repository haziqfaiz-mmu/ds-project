Dataset 2 : College and Universities in the US

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Colleges_and_Universities.csv")

df = df[["IPEDSID","NAME","ADDRESS","CITY","STATE","ZIP","TYPE","COUNTY","LATITUDE","LONGITUDE","PT_ENROLL","FT_ENROLL","TOT_ENROLL","WEBSITE"]]
df
```

```python
#1 = public
#2 = private non-profit
#3 = private for profit
def renameType(i):
    if(i==1):
        i= "public"
    elif(i==2):
        i = "private non-profit"
    elif(i==3):
        i = "private for profit"
    return i
    
df["TYPE"]=df["TYPE"].apply(renameType)
df["TYPE"]
```

```python
#remove null
df=df.drop(df[df['TOT_ENROLL'] ==0].index)
df
```
