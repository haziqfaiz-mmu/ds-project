Dataset 1: UNICEF Primary School Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("fusion_EDUCATION_UNICEF_1.0_all.csv")
df
```

```python
df=df.drop(columns=["DATAFLOW"])
df = df[["REF_AREA:Geographic area","INDICATOR:Indicator","SEX:Sex","EDUCATION_LEVEL:Education Level","WEALTH_QUINTILE:Wealth Quintile","RESIDENCE:Residence","OBS_VALUE:Observation Value","TIME_PERIOD:Time period"]]
df
```

```python
new_headers = ["Country","Indicator","Sex","Education Level", "Wealth Quintile","Residence","Value","Time"]
```

```python
df.columns = new_headers
df
```

```python
primary_schools = df[df["Education Level"]=="ISCED11_1: Primary education"]
totaldf = df[df["Education Level"]=="_T: Total"]
primary_schools.head()
totaldf.head()
```

```python
#Correct the name
def correctName(string):
    string = string[5:]
    return string

primary_schools["Country"]=primary_schools["Country"].apply(correctName)
primary_schools.head()
```

```python
#Correct Indicator
def correctIndicator(string):
    string = string[12:]
    return string

primary_schools["Indicator"]=primary_schools["Indicator"].apply(correctIndicator)
primary_schools.head()
```

```python
#Correct Sex
def correctSex(string):
    string = string[3:]
    return string

primary_schools["Sex"]=primary_schools["Sex"].apply(correctSex)
primary_schools.head()
```

```python
#Correct Education Level
def correctEducation(string):
    string = string[11:]
    return string

primary_schools["Education Level"]=primary_schools["Education Level"].apply(correctEducation)
primary_schools.head()
```

```python
#Correct Wealth_Quintile
def correctWealth(string):
    string = string[4:]
    return string

primary_schools["Wealth Quintile"]=primary_schools["Wealth Quintile"].apply(correctWealth)
primary_schools.head()
```

```python
#Correct Residence
def correctResidence(string):
    string = string[4:]
    return string

primary_schools["Residence"]=primary_schools["Residence"].apply(correctResidence)
primary_schools.head()
```

```python
primary_schools
```

```python
primary_schools.to_csv("cleaned_primary_school.csv",index = False)
```

```python
cleandf = pd.read_csv("cleaned_primary_school.csv")
cleandf
```

```python
gdpdf = pd.read_csv("Country_GDP_PPP.csv")
gdpdf = gdpdf[["LOCATION","Value","TIME"]]
gdpdf=gdpdf.rename({"LOCATION":"Country","TIME":"Time", "Value":"GDP"},axis=1)
gdpdf
```

```python
import pycountry

country = pycountry.countries.get(alpha_3="AUS")
country_name = country.name
print(country_name)

def convertCountry(country_code):
    if(pycountry.countries.get(alpha_3=country_code)):
        country = pycountry.countries.get(alpha_3=country_code)
        country_name = country.name
        return country_name
    else:
        return country_code
    
gdpdf["Country"] = gdpdf["Country"].apply(convertCountry)
gdpdf
```

```python
mergeddf = pd.merge(cleandf,gdpdf,on =["Country","Time"])
```

```python
#remove whitespace
for column in mergeddf:
    if mergeddf[column].dtypes == "object":
        mergeddf[column] = mergeddf[column].str.strip()
mergeddf
```

```python
mergeddf.to_csv("cleaned_primary_school_GDP.csv",index = False)
```
