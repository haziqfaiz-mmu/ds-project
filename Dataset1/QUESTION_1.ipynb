```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("cleaned_primary_school_GDP.csv")
df
```

```python
#Which wealth quintile has the highest completion rate for primary education in each country?


df2= df[df["Indicator"]=="Completion rate for children of primary school age"]
df3 = df.set_index("Wealth Quintile")
df3 = df3.drop("Total")


df4 = df3.groupby(['Country','Wealth Quintile']).sum()
df4[['Value']]
```

```python
letter_pivot = pd.pivot_table(
    df3, index='Country', columns='Wealth Quintile', values='Value', aggfunc='sum'
)
#letter_pivot

letter_pivot.plot.barh(figsize=(15, 15))
```
