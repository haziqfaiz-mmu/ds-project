Dataset 1: UNICEF Primary School Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("fusion_EDUCATION_UNICEF_1.0_all.csv")
df
```

```python
df=df.drop(columns=["DATAFLOW"])
df = df[["REF_AREA:Geographic area","INDICATOR:Indicator","SEX:Sex","EDUCATION_LEVEL:Education Level","WEALTH_QUINTILE:Wealth Quintile","RESIDENCE:Residence","OBS_VALUE:Observation Value"]]
df
```

```python
new_headers = ["Geographic Area","Indicator","Sex","Education Level", "Wealth Quintile","Residence","Value"]
df.columns = new_headers
df.head()
```

```python
primary_schools = df[df["Education Level"]=="ISCED11_1: Primary education"]
totaldf = df[df["Education Level"]=="_T: Total"]
primary_schools.head()
totaldf.head()
```

```python
def correctName(string):
    string = string[5:]
    return string

primary_schools["Geographic Area"]=primary_schools["Geographic Area"].apply(correctName)
primary_schools.head()
```
